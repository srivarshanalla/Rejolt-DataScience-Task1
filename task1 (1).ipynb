```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("heart.csv")
data.head
print(data.isnull())
```

```python
data.head(8)
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
#total missing values
total = np.product(data.shape)
a=data.isnull()
t_missing = a.sum()

#percentage
(t_missing/total)*100
```

```python
data.fillna(method='bfill',axis=0).fillna(0)
```

```python
duplicate_rows=data.duplicated()
print("duplicated rows:",duplicate_rows.sum())
```

```python
data.drop_duplicates(inplace=True)
```

```python
duplicate_rows=data.duplicated()
print("duplicated rows:",duplicate_rows.sum())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 7)) 
x = data["Age"] 
  
plt.hist(x, bins = 20, color = "green") 
plt.title("Count of cholestrol at particular age") 
plt.xlabel("Age") 
plt.ylabel("Cholestrol")
```
